#### QUESTION: Why were there a lots of zero values in the data set
#### Observations: 
#### 1. There seems to be many rows for New_cases and New_deaths  with negative values.
#### 2. All the negative values do not deviate from their neighbouring rows.
#### Conclusions: Maybe as a result typographical error, therefore deleting the rows will cause losing considerable amount of information. .
##### The below few lines of code below fetches the indexes of the negative values and then taking their absolute zeroes respectively. Same will be done to the New_deaths row of zero values

```python
world_covid_df[world_covid_df.New_deaths < 0]
```

```python
world_covid_df[world_covid_df.New_cases < 0]
```

```python
for index in (world_covid_df[world_covid_df.New_cases < 0]).index:
    world_covid_df.at[index, 'New_cases'] = abs(world_covid_df.at[index, 'New_cases'])
for index in (world_covid_df[world_covid_df.New_deaths < 0]).index:
    world_covid_df.at[index, 'New_deaths'] = abs(world_covid_df.at[index, 'New_deaths'])
    
```
